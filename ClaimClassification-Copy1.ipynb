```python
# Separate majority and minority classes
df_majority = df1[df1['Is_Business_Methods'] == False]
df_minority = df1[df1['Is_Business_Methods'] == True]
```

```python
# Upsample minority class
df_minority_upsampled = resample(df_minority, replace=True, n_samples=397, random_state=123) 
```

```python
# Combine majority class with upsampled minority class
df_upsampled = pd.concat([df_majority, df_minority_upsampled])
```
